The format used for date is SYYMMDDhhmmss, and it stands for
Season, Year, Month, Day, hour, minute, second

Import pandas library in order to work with data frame using Python

```python
import pandas

df = pandas.DataFrame({
    'Date': ["H170101000000", "H170101000001", "H170101000000", "H170101000000", "H170101000002"],
    'User_ID':[1, 1, 2, 2, 1],
    'Consommation': [105, 100, 250, 250, 110]
})

df = df[['Date','User_ID','Consommation']]
df.head(5)
```

```python
df_organized = df.copy()
df_organized['Year'] = df_organized['Date'].str[1:3]
df_organized['Season'] = df_organized['Date'].str[0]
df_organized['Month'] = df_organized['Date'].str[3:5]
df_organized['Day'] = df_organized['Date'].str[5:7]
df_organized['Hour'] = df_organized['Date'].str[7:9]
df_organized['Minute'] = df_organized['Date'].str[9:11]
df_organized['Second'] = df_organized['Date'].str[11:13]

df_organized.head(5)
```

```python
import matplotlib.pyplot as plt

consommation = df_organized['Consommation'][df_organized['User_ID'] == 1].values.tolist()
seconds = df_organized['Second'][df_organized['User_ID'] == 1].values.tolist()

plt.plot(seconds, consommation)
plt.show()
```
